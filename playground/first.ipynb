```python
import pandas as pd
import numpy as np
```

```python
langen_path = 'EnglishTranslation/EnglishTranslation-master/'
langtw_path = 'TOSTW-Lang/'
```

```python
def read_langfile(filepath, filename):
    df = pd.read_csv(filepath + filename, sep='\t', names=['no', 'name', 'korean'], usecols=[0,1,2])
    return df.iloc[:, :3]
```

```python
def match_langfiles(filename):
    df_en = read_langfile(langen_path, filename)
    df_tw = read_langfile(langtw_path, filename)
    
    
```

```python
df1 = read_langfile(langen_path, 'ITEM.tsv')
df1 = df1.iloc[:, :3]
df1.head()
```

```python
df1.tail()
```

```python
df2 = read_langfile(langtw_path, 'ITEM.tsv')
df2.head()
```

```python
def mergeprocess(df1, df2):
    df = pd.merge(df1, df2, on='korean', how='left')
    df['name_y2'] = np.where(df['name_y'].isna(), df['name_x'], df['name_y'])
    return df
    
df_merge = mergeprocess(df1, df2)
df_merge.head()
```

```python
df_itos_tw = df_merge[['no_x', 'name_y2']]
df_itos_tw.head()
```

```python
df_itos_tw.to_csv('itos-tw/' + 'ITEM.tsv', sep='\t', index=False, header=False)
```
